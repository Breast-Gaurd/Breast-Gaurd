#### Splitting the dataset into dependent and independent variables:

```python
# x ---> Independent variables
x=data.iloc[:,1:]
x
```

```python
# y ---> Dependent variable
y=data.diagnosis
y
```

#### Encoding:

Diagnosis is the only categorical column. So, performing Label Encoding on 'diagnosis' column.

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lr=LabelEncoder()
```

```python
y=lr.fit_transform(y)
y
```

#### Scaling the independent variables using Normalization:

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
x_scale=scaler.fit_transform(x)
x_scale
```

```python
name=x.columns
x=pd.DataFrame(x_scale,columns=name)
x
```

#### Spliting the dataset into Training and Testing Data:

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=14)
```

#### Model Building:

```python
# SUPPORT VECTOR MACHINE
from sklearn.svm import SVC
model1=SVC(kernel='linear')
# Model Training:
msvc=model1.fit(x_train,y_train)
```

```python
# RANDOM FOREST
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)
# Model Training
rf.fit(x_train,y_train)
```

```python
# NAIVE BAYES CLASSIFIER
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
# Model Training
nb.fit(x_train,y_train)
```

```python
# DECISION TREE
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
# Model Training
dt.fit(x_train,y_train)
```

```python
# Model Testing for Support Vector Machine:
pred_svm=msvc.predict(x_test)
#print(pred_svm)

# Model Testing for Random Forest:
pred_rf=rf.predict(x_test)
#print(pred_rf)

# Model Testing for Naive Bayes Classifier:
pred_nb=nb.predict(x_test)
#print(pred_nb)

# Model Testing for Decision Tree:
pred_dt=dt.predict(x_test)
#print(pred_dt)
```

#### Evaluating the performance of the models:

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report,precision_score
```

```python
print('Accuracy Precision score of SVM:')
print(accuracy_score(pred_svm,y_test))
print('Precision score of SVM:')
print(precision_score(pred_svm,y_test))

print('\n'+'Accuracy score of Random Forest Model:')
print(accuracy_score(pred_rf,y_test))
print('Precision score of Random Forest Model:')
print(precision_score(pred_rf,y_test))

print('\n'+'Accuracy Precision score of Naive Bayes Classifier:')
print(accuracy_score(pred_nb,y_test))
print('Precision score of Navie Bayes Classifier:')
print(precision_score(pred_nb,y_test))

print('\n'+'Accuracy Precision score of Decision Tree:')
print(accuracy_score(pred_dt,y_test))
print('Precision score of Decision Tress:')
print(precision_score(pred_dt,y_test))
```

```python
print('Confusion Matrix of SVM Model: ')
sns.heatmap(confusion_matrix(pred_svm,y_test),annot=True,fmt='g',
            xticklabels=['malignant', 'benign'],
            yticklabels=['malignant', 'benign'])
plt.ylabel('Prediction',fontsize=13)
plt.xlabel('Actual',fontsize=13)
plt.title('Confusion Matrix',fontsize=17)
plt.show()
```
