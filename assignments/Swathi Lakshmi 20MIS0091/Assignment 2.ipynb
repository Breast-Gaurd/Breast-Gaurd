### 5. Handle the missing values.

```python
# Checking for null values
data.isnull().any()
```

```python
# Filling the missing values in age column using mean, because age is a numerical value
data['age'].fillna(data['age'].median(),inplace=True)
```

```python
# Filling the missing values in embarked column, because embarked is categorical
data['embarked'].fillna(data['embarked'].mode()[0],inplace=True)
```

```python
# Filling the missing values in deck column, because deck is categorical
data['deck'].fillna(data['deck'].mode()[0],inplace=True)
```

```python
# Filling the missing values in embark_town column, because embark_town is categorical
data['embark_town'].fillna(data['embark_town'].mode()[0],inplace=True)
```

```python
# Checking for null values after imputation
data.isnull().any()
```

```python
# Thus, the imputation is successful.
```
