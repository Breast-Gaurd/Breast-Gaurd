### 6. Find the outliers and replace the outliers.

```python
# Using figure to plot the box plots for the numerical columns to visualize the outliers.

plt.figure(figsize=(12,8))

#price column
plt.subplot(2,3,1)
sns.boxplot(data.price)
plt.title('Price')

#area column
plt.subplot(2,3,2)
sns.boxplot(data.area)
plt.title('Area')

#bedrooms column
plt.subplot(2,3,3)
sns.boxplot(data.bedrooms)
plt.title('Bedrooms')

#bathrooms column
plt.subplot(2,3,4)
sns.boxplot(data.bathrooms)
plt.title('Bathrooms')

#stories column
plt.subplot(2,3,5)
sns.boxplot(data.stories)
plt.title('Stories')

#parking column
plt.subplot(2,3,6)
sns.boxplot(data.parking)
plt.title('Parking')
```

```python
# Removing autliers

# Price column
data=data[data.price<9205000]  # 9095625 is the upper extreme of the price column

# Area column
data=data[data.area<10500]  # 10500 is the upper extreme of the area column
```

```python
# Visualizing the outliers after treating them.

plt.figure(figsize=(10,4))

#price column
plt.subplot(1,2,1)
sns.boxplot(y='price', data=data)

#area column
plt.subplot(1,2,2)
sns.boxplot(y='area', data=data)
```

```python
# Label encoding for mainroad column
data.mainroad=le.fit_transform(data.mainroad)

# Label encoding for guestroom column
data.guestroom=le.fit_transform(data.guestroom)

# Label encoding for basement column
data.basement=le.fit_transform(data.basement)

# Label encoding for hotwaterheating column
data.hotwaterheating=le.fit_transform(data.hotwaterheating)

# Label encoding for airconditioning column
data.airconditioning=le.fit_transform(data.airconditioning)

# Label encoding for furnishingstatus column
data.furnishingstatus=le.fit_transform(data.furnishingstatus)
```

### 8. Split the data into dependent and independent variables.

```python
# Y is the dependent variable - "price" class label

Y=data['price']
Y.head()
```

```python
# X is the independent variables 

X=data.drop(columns=['price'],axis=1)
X.head()
```

### 9. Scale the independent variables.

```python
# Using MinMax scaling technique:

from sklearn.preprocessing import MinMaxScaler
scale=MinMaxScaler()
X_scaled=scale.fit_transform(X)
X_scaled
```

```python
name=X.columns
print(name)
```

```python
# printing Scaled independent variables

X=pd.DataFrame(X_scaled,columns=name)
X    
```

### 10. Split the data into training and testing.

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

### 11. Build the model.

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
```

### 12. Train the model.

```python
lr.fit(X_train,Y_train)
```

### 13. Test the model.

```python
# Prediction
Y_pred=lr.predict(X_test)
Y_pred                    
```

```python
# Actual Outcome
Y_test
```

### 14. Measure the performance using metrics.

```python
# Finding the ERROR

E=Y_test-Y_pred
E
```

```python
# R2 score

from sklearn.metrics import r2_score
acc=r2_score(Y_test,Y_pred)
print('Accuracy percentage: ')
print(acc*100)
```
